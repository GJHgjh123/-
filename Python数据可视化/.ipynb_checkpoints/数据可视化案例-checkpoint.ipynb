```python
import numpy as np
x=np.arange(0,1,0.05)
print(x)
```

```python
#y=sin(2*pi*x)
y=np.sin(2*np.pi*x)
print(y)
```

```python
plt.plot(x,y,'b--*',label='sin')
plt.title('My First Plot')
plt.xlabel('x ')
plt.ylabel('y ')
plt.legend()
plt.show()
```

```python
fig,ax=plt.subplots(2,2)
ax[1,1].plot(x,y)
plt.show()
```

```python
#y2=np.cos(2*pi*x)
y2=np.cos(2*np.pi*x)
print(y2)
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,'b--*',label='sin')
ax.plot(x,y2,'r--o',label='cos')
ax.set(title='My Plot')
legend=ax.legend(loc='best')
plt.show()
```

```python
#散点图及三维散点图
import numpy as np
data=np.random.randint(0,100,size=[40,40])
data
```

```python
x,y=data[0],data[1]
ax=plt.subplot(111)
ax.scatter(x[:10],y[:10],c='r')
ax.scatter(x[10:20],y[10:20],c='b')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

```python
#三维散点图
from mpl_toolkits.mplot3d import Axes3D

x,y,z=data[0],data[1],data[2]
ax=plt.subplot(111,projection='3d')
ax.scatter(x[:10],y[:10],z[:10],c='b')
ax.scatter(x[10:20],y[10:20],z[10:20],c='r')
ax.scatter(x[30:40],y[30:40],z[30:40],c='g')

ax.set_xlabel('X')
ax.set_zlabel('Z')
ax.set_ylabel('Y')

plt.show()
```

```python
#散点图矩阵、获取sklearn包里面的鸢尾花数据集
from sklearn.datasets import load_iris
iris_dataset=load_iris()
```

```python
#print(iris_dataset)
#print(iris_dataset.feature_names)
print(iris_dataset['data'][:5])
print(iris_dataset['feature_names'][:5])
#iris_dataset[sepal width (cm)][:5]
#iris_dataset
```

```python
#绘制散点图矩阵之表格化
import pandas as pd
#1)得到数据
iris_df=pd.DataFrame(iris_dataset['data'],columns=iris_dataset.feature_names)
iris_df.head()
```

```python
#鸢尾花分类API
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
# 2）划分数据集
x_train, x_test, y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target, random_state=0)
# 3）特征工程：标准化
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
# 4）KNN算法预估器
estimator = KNeighborsClassifier(n_neighbors=3)
estimator.fit(x_train, y_train)
# 5）模型评估
#方法：计算准确率
score = estimator.score(x_test, y_test)
print("准确率为：\n", score)
y_predict = estimator.predict(x_test)
iris_df1=pd.DataFrame(x_test,columns=iris_dataset.feature_names)
grr=pd.plotting.scatter_matrix(iris_df1,c=y_test,figsize=(15,15),marker='o',s=60,alpha=0.8,cmap=mglearn.cm3)
grr1=pd.plotting.scatter_matrix(iris_df,c=iris_dataset['target'],figsize=(15,15),marker='o',s=60,alpha=0.8,cmap=mglearn.cm3)
```

```python
#逻辑回归中的sigmoid函数
#matplotlib,numpy
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
#第一种画图方法
'''x=np.arange(-5,6,1)
fig,ax=plt.subplots()
ax.plot(x,sigmoid(x),'r')'''
#第二种
import matplotlib.pyplot as plt
x=np.arange(-5,6,1)
plt.plot(x,sigmoid(x),'r--*')
plt.title('sigmoid函数图像')
plt.xlabel('x ')
plt.ylabel('y ')
plt.show()
```
